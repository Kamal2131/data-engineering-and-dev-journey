### Working with Text Files

Text files stores plaintext like , logs, notes, or configurations.

Python provide buil-in `open()` function to work with them.

```python
with open("day4_data/myfile.txt", "w") as f:
    f.write("Day 4 of DataDev Mastery ")

with open("day4_data/myfile.txt", "r") as f:
    print(f.read())
```

#### Explanation

- "w" â†’ Write mode (creates or overwrites the file)

- "r" â†’ Read mode

- "a" â†’ Append mode (adds to existing content)

- "r+" â†’ Read and Write mode

- with open() â†’ Context manager, automatically closes the file (even if an error occurs)

#### Real-world Use

- Youâ€™ll use text file handling for:

- Logging system activity

- Storing temporary data

- Reading small configuration files

ðŸ” Pro Tip

For large files, donâ€™t use .read() directly â€” it loads the entire file into memory.

Use:

```python
for line in f:
    print(line.strip())
```

Or

```python
f.readline()    # Reads one line at a time
f.readlines()   # Reads all lines into a list
```

## 2. Working with CSV Files
#### ðŸ“Š What is CSV?

- A Comma-Separated Values (CSV) file stores tabular data (rows and columns), often exported from Excel, databases, or APIs.

#### Option 1: Using csv module

```python
import csv

with open("day4_data/students.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["Name", "Marks"])
    writer.writerow(["Kamal", 88])
```

#### Explanation:

- csv.writer() â†’ creates a writer object

- .writerow() â†’ writes a single row (list)

- newline="" â†’ prevents extra blank lines in Windows

##### To Read

```python
import csv

with open("day4_data/students.csv", "r") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
```

### Option 2: Using pandas

- When working with large datasets, pandas is faster and more flexible.

```python
import pandas as pd

df = pd.read_csv("day4_data/students.csv")
print(df.head())
```

### 3. Working with JSON Files
#### ðŸ“˜ What is JSON?

- JSON (JavaScript Object Notation) is the most common format for APIs and configurations.
- Itâ€™s structured like a Python dictionary: {key: value}.

##### Example

```python
import json

data = {"name": "Kamal", "skills": ["Python", "FastAPI", "Airflow"]}

with open("day4_data/profile.json", "w") as f:
    json.dump(data, f, indent=2)
```

- json.dump() â†’ writes dictionary data to a file

- indent=2 â†’ makes it readable

##### To read it back:

```python
with open("day4_data/profile.json", "r") as f:
    data = json.load(f)
    print(data)
```

#### JSON is Essential For:

- APIs (input/output data)

- Configurations (credentials, parameters)

- MongoDB or NoSQL data exchange

- ETL pipelines (structured data)

### 4. Error Handling in File Operations

#### When dealing with files, things can go wrong:

- File doesnâ€™t exist

- Permission denied

- Wrong format

#### Use try-except to handle these gracefully:

```python
try:
    with open("day4_data/student.csv", "r") as f:
        print(f.read())
except FileNotFoundError:
    print("File not found!")
except Exception as e:
    print("Something went wrong:", e)
```
