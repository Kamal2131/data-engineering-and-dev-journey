## 6. List Comprehension (important)

### Putting All the squared of number 1 - 6 without List Comprehension

```python
sq = []
for i in range(1, 6):
    sq.append(i**2)
sq
```

### Putting All the squared of number 1 - 6 with List Comprehension

```python
sq = [i**2 for i in range(1,6)]
sq
```

### Another Example

```python
even_num = [i for i in range(1,6) if i % 2 == 0]
even_num
```
